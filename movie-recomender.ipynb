```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits, on ='title')
```

```python
movies.head(1)
```

```python
# columns that are needed: genre, id, keywords, title, overview, cast, crew

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head(1)
```

```python
#missing data
movies.isnull().sum()
```

```python
movies.dropna(inplace= True)
```

```python
#duplicated data
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#converting genres string of list of dictionaries into list of dicts
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): #string to list conversion
        L.append(i['name'])         #'name' attributed is needed
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
#now for keywords column

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
#now for cast, only first 3 actors names needed

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):    #string to list conversion
        if counter != 3:
            L.append(i['name'])         #'name' attributed is needed
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
#for crew where job = director, (only director name needed)

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(1) 
```

```python
#removing spaces between names 'Science Fiction' -> 'ScienceFiction'
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head(1)
```

```python
#concatenate all the columns to tags

movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head(1)
```

```python
#convert tag list into string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
#convert everything to lowercase

new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

```python
new_df['tags'][0]
```

```python
#convert tags to vectors: textvectorization (bag of words)
#largetext = concatenate all the tags: then find the 5000 most common words 
#                 w1   w2   w3 ..... w5000
#for movie 1, m1: 5    3    0          1    <- one vector of 5000 Dimensions
#for movie 2, m2: 3    7    1          0
#...
#matrix dataframe 5000x5000 (frequency of the most common words in each movie)

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
#apply stemming to make sure that loved, loving, lover into lov
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#distance between vectors will determine the similarity
#if distance increases, similarity decreases and vice versa
#cosine distance method

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
list(enumerate(similarity[0])) #created (index, similarti) tuples
```

```python
#recommend 5 similar movies using vectorization
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    #sorting on the basis of similarity in the tuple (index, similarity) from 1st pos to 6th pos
    distances = sorted(list(enumerate(similarity[movie_index])),reverse=True,key = lambda x: x[1])

    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
import pickle

pickle.dump(new_df,open('movie_list.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
```
